```python
# Aggregating Results
```

```python
%%sql

SELECT winners, count(winners) as Championships
FROM svjetski_kup
GROUP BY winners
ORDER BY Championships DESC
```

```python
# very smiple way to replicate this query in pandas using '.value_counts' method or '.groupby'
```

```python
svj_kup_df['Winners'].value_counts()
```

```python
svj_kup_df[['Year', 'Winners']].groupby('Winners')\
                               .count().reset_index()
```

```python
svj_kup_df[['Year', 'Winners']].groupby('Winners')\
                               .count()\
                               .rename(columns={'Year':'Championships'})\
                               .sort_values(by='Championships', ascending=False).reset_index()
```

```python
# Filtering aggregates with Having
# only the Couontries with more than 2 Championships
```

```python
%%sql

SELECT winners, count(winners) as Championships
FROM svjetski_kup
GROUP BY winners
HAVING Championships > 2        
ORDER BY Championships DESC
```

```python
# in PANDAS is even easyer by chain methond, where I use my last chain as base for 'query' method
```

```python
svj_kup_df[['Year', 'Winners']].groupby('Winners')\
                               .count()\
                               .rename(columns={'Year':'Championships'})\
                               .sort_values(by='Championships', ascending=False)\
                               .reset_index()\
                               .query('Championships > 2')
```
